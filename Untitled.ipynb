```python
import os
import torch
import torch.nn as nn
import torchvision.models as models
from torchvision import datasets, transforms
from torch.utils.data import DataLoader
import matplotlib.pyplot as plt

# -------------------------
# Device
# -------------------------
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

# -------------------------
# Dataset
# -------------------------
DATA_PATH = "Sorted"  # your dataset folder

transform = transforms.Compose([
    transforms.Resize((224, 224)),
    transforms.ToTensor(),
    transforms.Normalize(
        mean=[0.485, 0.456, 0.406],
        std=[0.229, 0.224, 0.225]
    )
])

test_dataset = datasets.ImageFolder(os.path.join(DATA_PATH, "test"), transform=transform)
test_loader = DataLoader(test_dataset, batch_size=32, shuffle=False)

# -------------------------
# Model
# -------------------------
class MalwareClassifier(nn.Module):
    def __init__(self, n_classes):
        super().__init__()
        self.resnet = models.resnet50(weights='DEFAULT')
        for p in self.resnet.parameters():
            p.requires_grad = False
        num_features = self.resnet.fc.in_features
        self.resnet.fc = nn.Sequential(
            nn.Linear(num_features, 1000),
            nn.ReLU(),
            nn.Linear(1000, n_classes)
        )
    def forward(self, x):
        return self.resnet(x)

n_classes = len(test_dataset.classes)
model = MalwareClassifier(n_classes)
model.load_state_dict(torch.load("malware_classifier.pth", map_location=device))
model.to(device)
model.eval()

# -------------------------
# FGSM attack
# -------------------------
def fgsm_attack(model, images, labels, epsilon=0.1):
    images = images.clone().detach().to(device)
    labels = labels.to(device)
    images.requires_grad = True

    outputs = model(images)
    loss = nn.CrossEntropyLoss()(outputs, labels)
    model.zero_grad()
    loss.backward()
    
    grad = images.grad.sign()
    adv_images = images + epsilon * grad
    adv_images = torch.clamp(adv_images, 0, 1)
    return adv_images.detach()

# -------------------------
# Visualization
# -------------------------
def imshow(tensor_img, title=None):
    if tensor_img.dim() == 4:
        tensor_img = tensor_img.squeeze(0)
    inv_normalize = transforms.Normalize(
        mean=[-0.485/0.229, -0.456/0.224, -0.406/0.225],
        std=[1/0.229, 1/0.224, 1/0.225]
    )
    img = inv_normalize(tensor_img).clamp(0, 1)
    img = img.permute(1, 2, 0).cpu().numpy()
    plt.imshow(img)
    if title:
        plt.title(title)
    plt.axis("off")

# -------------------------
# Pick an image and attack
# -------------------------
idx = 10  # change index here
image, label = test_dataset[idx]
image = image.unsqueeze(0).to(device)
label = torch.tensor([label]).to(device)

with torch.no_grad():
    pred = model(image).argmax(dim=1)

print(f"True label: {test_dataset.classes[label.item()]}")
print(f"Predicted before FGSM: {test_dataset.classes[pred.item()]}")

x_adv = fgsm_attack(model, image, label, epsilon=0.1)

with torch.no_grad():
    adv_pred = model(x_adv).argmax(dim=1)

print(f"Predicted after FGSM: {test_dataset.classes[adv_pred.item()]}")

plt.figure(figsize=(8,4))
plt.subplot(1,2,1)
imshow(image, title=f"Original\n{test_dataset.classes[pred.item()]}")
plt.subplot(1,2,2)
imshow(x_adv, title=f"Adversarial\n{test_dataset.classes[adv_pred.item()]}")
plt.show()
```

```python
import torch

# =========================
# Candidate epsilon values
# =========================
eps_candidates = [0.05, 0.1, 0.2]  # adjust based on your malware image scale
success_image, success_label, success_eps = None, None, None
# If you already have the class definition in this session, you can do:
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model_path = "malware_classifier.pth"

model = torch.jit.load(model_path, map_location=device)
model.eval()
model.to(device)

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

# If you already have the class definition in this session, you can do:
model_path = "malware_classifier.pth"
model = torch.jit.load(model_path, map_location=device)
model.eval()
model.to(device)

# =========================
# Pick a correctly classified example from test set
# =========================
target_class_name = "worm"  # class you want to attack from
target_class_idx = test_dataset.class_to_idx[target_class_name]

# Pick a target label to attack (targeted attack)
attack_target_name = "trojan"  # class we want the model to misclassify as
attack_target_idx = test_dataset.class_to_idx[attack_target_name]

for xb, yb in test_loader:
    xb, yb = xb.to(device), yb.to(device)
    
    # Find indices of the chosen source class
    match_indices = (yb == target_class_idx).nonzero(as_tuple=True)[0]
    if len(match_indices) == 0:
        continue

    # Check predictions for this class
    with torch.no_grad():
        preds = model(xb[match_indices]).argmax(dim=1)
        correct_mask = (preds == target_class_idx)
        if correct_mask.any():
            # Take first correctly classified sample
            local_idx = correct_mask.nonzero(as_tuple=True)[0][0].item()
            idx = match_indices[local_idx].item()
            candidate = xb[idx]
            candidate_label = yb[idx]
            break

if candidate is None:
    raise RuntimeError(f"Could not find a correctly classified '{target_class_name}' in test set")

# =========================
# Run targeted FGSM attack
# =========================
target_label = torch.tensor([attack_target_idx], device=device)

for eps_try in eps_candidates:
    x_adv = fgsm_attack(
        model,
        candidate.unsqueeze(0),
        target_label,
        epsilon=eps_try,
        targeted=True
    )
    with torch.no_grad():
        pred = model(x_adv).argmax(dim=1).item()
    print(f"epsilon={eps_try:.3f} -> predicted {test_dataset.classes[pred]}")

    if pred == attack_target_idx:
        success_image = candidate
        success_label = candidate_label
        success_eps = eps_try
        break

if success_image is not None:
    print(f"\nSuccess! Epsilon={success_eps:.3f} fooled the model from "
          f"{test_dataset.classes[success_label.item()]} -> {attack_target_name}")
else:
    print("\nAttack unsuccessful for given epsilon candidates.")
```
